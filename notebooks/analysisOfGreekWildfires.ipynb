### Δασικές πυρκαγιές στην Ελλάδα από το 2000 έως σήμερα: Μία στατιστική προσέγγιση

## Abstract

Το καλοκαίρι του 2021, και ειδικά ο Αύγουστος, αποτέλεσαν μία πολύ δύσκολη περίοδο για την Ελλάδα κατά την οποία μεγάλα πύρινα μέτωπα κατέκαψαν πάνω από 1.250.000 στρέμματα γης. Ένας παράγοντας για τον οποίο επικρίθηκε ο κρατικός μηχανισμός είναι η αποτυχία του να προλάβει τις φωτιές όσο ήταν μικρές. Η τοποθέτηση των υπεύθυνων για ζητήματα πολιτκής προσταστίας, από πλευράς του κράτους ήταν πως οι ακραίες καιρικές συνθήκες του Αυγούστου μαζί με τον πολύ μεγάλο αριθμό πυρκαγιών που εμφανίζονταν κάθε μέρα αποτελούν τις βασικές αιτίες για την καταστροφή που σημειώθηκε. Η παρούσα στατιστική μελέτη έχει ως σκοπό να διερευνήσει τα δημόσια διαθέσιμα στατιστικά που παρέχει το Πυροσβεστική Σώμα (εφεξής ΠΣ), για τις δασικές πυρκαγιές για τα έτη 2000-2021. Τα δεδομένα αυτά αποτελούν καταγραφή όλων των συμβάντων στα οποία ανταποκρίνεται το ΠΣ και περιέχουν δεδομένα ως προς τον τόπο και χρόνο εκδήλωσης της κάθε πυρκαγιάς, την ζημία που αυτή προκάλεσε καθώς και το δυναμικό του ΠΣ που συμμετείχε στην κατάσβεση (το δεδομένο αυτό διατίθεται μόνο από το 2013 και μετά). Τα δεδομένα αυτά θα μπορούσαν να δώσουν μια εικόνα σε μία σειρά από ερωτήματα, που τέθηκαν στο δημόσιο διάλογο, όπως το γιατί δεν σβήστηκαν έγκυρα οι φετινές πυρκαγιές,

Τα δεδομένα διατίθενται από το ΠΣ: [σε μορφή αρχέιων Excel](https://www.fireservice.gr/el_GR/synola-dedomenon) και σε μια πιο [ευαναγνωστη μορφή με πίνακες](https://www.fireservice.gr/el_GR/stoicheia-symbanton) και στο [ετήσιο έντυπο δραστηριοτήτων](https://www.fireservice.gr/el_GR/drasteriotetes-p.s.)


```python
import os
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
# Loading the data
df_list = list()
year_list = ["2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]
file_list = list()
directory = os.getcwd().replace('notebooks', 'data/')
# Sort excels by year
for filename in os.listdir(directory):
    file_list.append(filename)
file_list.sort()
print(file_list)

# Load excels in dataFrame
for excel in file_list:
    print("Loading... {}".format(excel))
    if excel.endswith(".xls") or excel.endswith(".xlsx"):
        if excel.endswith("Dasikes_Pyrkagies_2000-2012.xls"):
            for i in range(0,13):
                df_list.append(pd.read_excel(directory + excel, i))
        else:
            df_list.append(pd.read_excel(directory + excel))
    print("Loaded: {}".format(excel))
# print(df_list) # print loaded data
```

```python
#Forest wildfires per year
fires_per_year = list()
for dictionary in df_list:
    fires_per_year.append(len(dictionary.index) - 1)
print(fires_per_year) # Print exact values
y_pos = np.arange(len(fires_per_year))

plot.rcdefaults()
plot.rcParams["figure.figsize"] = (20,7)
fig, ax = plot.subplots()



ax.barh(y_pos, fires_per_year, align='center', color='red')
ax.set_ylabel("Έτος")
ax.set_yticks(y_pos)
ax.set_yticklabels(year_list)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Αριθμός Δασικών Πυρκαγιών')
ax.set_title('Δασικές πυρκαγιές στην Ελλάδα ανα έτος (2000-2021)')
for i, value in enumerate(fires_per_year):
    ax.text(value + 100, i, str(value), color='black', fontweight='bold', fontsize=11, ha='left', va='center')

plot.show()
```
